```python
import pandas as pd

data = pd.read_csv("data/spotify.csv", index_col=0)
data = data[['artists','name','acousticness', 'danceability', 'duration_ms', 
             'energy', 'speechiness', 'instrumentalness', 'liveness', 'loudness', 
             'valence', 'tempo', 'popularity', 'explicit', 'mode', 'year']]
data
```

```python
from ast import literal_eval

# Keep only songs that have one and only one artist.
data.artists = [literal_eval(s) for s in data.artists]
data = data[data.artists.apply(len) == 1]
data.artists = [l[0] for l in data.artists]

data
```

```python
# top artists with the most songs in our dataset
top30 = data.artists.value_counts().iloc[:30]
top30
```

```python
data = data[data.artists.isin(top30.index)]
```

```python
# standardize data
quantitative_vars = data.select_dtypes('number').columns
quantitative_data = data.loc[:, quantitative_vars]
quantitative_data = (quantitative_data - quantitative_data.mean()) / quantitative_data.std()
data.loc[:, quantitative_vars] = quantitative_data
data
```

```python
from qclustering import dissimilarity_matrix

dmatrix = dissimilarity_matrix(data, 'artists')
dmatrix
```

```python
from qclustering import plot_dissimilarity_2d

_, var = plot_dissimilarity_2d(data, 'artists', figsize=(14,12))
```

Percentage of variance (from the dissimilarity matrix) that is observable in 2 dimensions:

```python
print("{:.2%}".format(var))
```
